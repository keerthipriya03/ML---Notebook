# Hyperparameter Tuning:
    GridSearchCV
    RandomizedSearchCV

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
# url = "https://raw.githubusercontent.com/datasets/breast-cancer/main/data/breast-cancer.csv"
# breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
url = "https://raw.githubusercontent.com/datasets/breast-cancer/main/data/breast-cancer.csv"
breast_cancer_dataset = pd.read_csv(url)
```

```python
print(breast_cancer_dataset)
```

```python
breast_cancer_dataset.head()
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)

data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# number of rows and Columns in this dataset
print(data_frame.shape)
# checking for missing values
print(data_frame.isnull().sum())
```

```python
# checking the distribution of Target Varibale
data_frame['label'].value_counts()
```

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
print(X , Y)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

# GridSearchCV
GridSearchCV is used for determining the best parameters for our model

```python
# loading the SVC model
model = SVC()
```

```python
# hyperparameters

parameters = {
              'kernel':['linear','poly','rbf','sigmoid'],
              'C':[1, 5, 10, 20]
}
```

```python
# grid search
classifier = GridSearchCV(model, parameters, cv=5)
```

```python
# fitting the data to our model
classifier.fit(X, Y)
```

```python
classifier.cv_results_
```

```python
# best parameters

best_parameters = classifier.best_params_
print(best_parameters)
```

```python
# higest accuracy

highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
# loading the results to pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score']]
grid_search_result
```

#RandomizedSearchCV

```python
# loading the SVC model
model = SVC()
```

```python
# hyperparameters

parameters = {
              'kernel':['linear','poly','rbf','sigmoid'],
              'C':[1, 5, 10, 20]
}
```

```python
# grid search
classifier = RandomizedSearchCV(model, parameters, cv=5)
```

```python
# fitting the data to our model
classifier.fit(X, Y)
```

```python
classifier.cv_results_
```

```python
# best parameters

best_parameters = classifier.best_params_
print(best_parameters)
```

```python
# higest accuracy

highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
# loading the results to pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
print(result.head)
print(result.head())
```

```python
randomized_search_result = result[['param_C','param_kernel','mean_test_score']]
print(randomized_search_result)
```
