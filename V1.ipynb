Stemming:

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content
                       if not word in stopwords.words('english')] #Remove common meaningless words
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
# convert the textual data to Feature Vectors
vectorizer = TfidfVectorizer()
```

```python
vectorizer.fit(X)

X = vectorizer.transform(X)
```

**vectorizer.fit(X)**
X is your array of cleaned text documents.
.fit() learns the vocabulary from all documents in X.
It calculates term frequencies (or TF-IDF values if using TfidfVectorizer).
This does not yet convert text to numbers; it only builds the internal mapping of words → feature indices.
**ex:**If X contains:["break news market", "polit debate heat"]
Vocabulary learned might be:["break", "news", "market", "polit", "debate", "heat"]

**X = vectorizer.transform(X)**
Converts each text document in X into a numerical vector.
Each vector shows the TF-IDF score (or count) of each word in the vocabulary.
The result is a sparse matrix of shape (num_documents, num_features).
**Example:**
Text → Vector:
"break news market" → [1, 1, 1, 0, 0, 0]
"polit debate heat" → [0, 0, 0, 1, 1, 1]


.fit() → learns vocabulary and statistics (TF, IDF, etc.)

.transform() → applies the learned vocabulary to convert text into numbers
